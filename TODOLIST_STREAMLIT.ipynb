```python
import streamlit as st

class TodoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append({"task": task, "completed": False})

    def delete_task(self, index):
        if 0 <= index < len(self.tasks):
            del self.tasks[index]
            st.write("Task deleted successfully.")
        else:
            st.write("Invalid task index.")

    def mark_completed(self, index):

        if 0 <= index < len(self.tasks):
            self.tasks[index]["completed"] = True
            st.write("Task marked as completed.")
        else:
            st.write("Invalid task index.")

    def display_tasks(self):
        if self.tasks:
            st.write("Tasks:")
            for i, task in enumerate(self.tasks):
                status = "Completed" if task["completed"] else "Pending"
                st.write(f"{i + 1}. {task['task']} - {status}")
        else:
            st.write("No tasks.")

def main():
    st.title("TODO List")
    todo_list = TodoList()

    option = st.selectbox("TODO List Options:", ["Add Task", "Delete Task", "Mark Task as Completed", "View Tasks"])

    if option == "Add Task":
        task = st.text_input("Enter task:")
        if st.button("Add"):
            todo_list.add_task(task)
    elif option == "Delete Task":
        index = st.number_input("Enter task index to delete:", value=1, step=1)
        if st.button("Delete"):
            todo_list.delete_task(int(index) - 1)
    elif option == "Mark Task as Completed":
        index = st.number_input("Enter task index to mark as completed:", value=1, step=1)
        if st.button("Mark Completed"):
            todo_list.mark_completed(int(index) - 1)
    elif option == "View Tasks":
        todo_list.display_tasks()

if __name__ == "__main__":
    main()
```
